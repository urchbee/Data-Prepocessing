# Data Preprocessing



Attached to this notebook is a dataset.

```python
import pandas as pd
```

1. Import the dataset into your notebook, print out the first 10 rows, check the statistical and general information

```python
url = 'https://raw.githubusercontent.com/urchbee/Data-Prepocessing/main/hepatitis.csv'
df1 = pd.read_csv(url)
#select first 10 rows by using head(num of columns/rows to print), if last 10 then use tail
df1.head(10)
```

```python
#describe is used to check the data for statitical info like mean 
df1.describe()
```

```python
#to check the general information on the data,use .info
df1.info()
```

2. Check for the sum of the missing values and the percentage of missing values.

```python
#sum of missing values- isnull().sum() 
missing_values = df1.isnull().sum()
missing_values
```

```python
#using len() to find number of rows in the data
dflen = len(df1)
dflen
#to find percentage of missing values, it would be the sum of missing values/lenght of rows
percentage = missing_values/dflen
percentage
```

3. Choose one option: Drop the columns with missing values or fill the columns with missing values. Give your reasons.

```python
#filling the missing columns with missing values using 'bfill' method which fills the dataset using data from the next row of the missing column
#this is because these columns are needed in this data analysis and also because the dataset isnt too bulky or clogged it can easily be filled
#each column indicates the testing done to ascertain hepatatis in each individual

hepatatis = df1.fillna(method = 'bfill')
hepatatis.head(10)
#use .head(number of rows to view) to get the output for the first number of rows
```

4. Choose a column and rename it. Give your reasons for picking that column

```python
#changing the column name from class to status using .rename method
#changing the last column name to status because it describes if alive/dead each individual in the dataset
new_hepatatis = hepatatis.rename(columns={"class" : "status"}) 

new_hepatatis.head(10)
```

5. Check the number of missing values after cleaning and state the impact on data analysis.

```python
#cleaning the data above by filling the missing values, the data is now less noisy with incomplete data
#also it has made the dataset more uniform and standardized which improves data quality 
new_hepatatis.isnull().sum()

#after the data is cleaned, the number of missing values is nil hence the dataset is uniform
```
